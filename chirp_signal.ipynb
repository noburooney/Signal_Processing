```
import numpy as np
import math
import random
import matplotlib.pyplot as plt

pi = math.pi
```

```
pnts = 1000
srate = 1024
t = np.arange(0, pnts)/srate
```

```
#bipolar chirp
freqmod = np.linspace(5, 15, pnts)

signal = np.sin(2*pi * (t + np.cumsum(freqmod)/srate))

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot(t, freqmod)

ax2 = fig.add_subplot(212)
ax2.plot(t, signal)
```

```
#multi polar chirp
k = 10
x = np.arange(0, k)
y = np.random.rand(k)

xvals = np.linspace(0, k-1, pnts)
freqmod_2 = 20 * np.interp(xvals, x, y)

signal_2 = np.sin(2*pi * (t + np.cumsum(freqmod_2)/srate))

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot(t, freqmod_2)

ax2 = fig.add_subplot(212)
ax2.plot(t, signal_2)

```
